```python
import tensorflow as tf
from tensorflow import keras
import os
import cv2
from sklearn.model_selection import train_test_split
import numpy as np
import matplotlib.pyplot as plt
from tensorflow.keras.preprocessing.image import ImageDataGenerator
from tensorflow.keras.utils import get_file
from tensorflow.keras.preprocessing.image import load_img, img_to_array
from tensorflow.keras.applications.resnet50 import preprocess_input, decode_predictions
import re
from tensorflow.keras.applications import ResNet50
from tensorflow.keras.layers import Dense
from tensorflow.keras.models import Model
from tensorflow.keras.applications import MobileNetV2
```

```python
print("Num GPUs Available: ", len(tf.config.list_physical_devices('GPU')))
```

```python
train_dir=r'D:\Studying\ApplAi_CV\Assignments\third\train'
validate_dir=r'D:\Studying\ApplAi_CV\Assignments\third\valid'
test_dir=r'D:\Studying\ApplAi_CV\Assignments\third\testFolder\test'
```

```python
train_imagenerator = ImageDataGenerator(
    
    rescale=1/255.,
    rotation_range=40,
    width_shift_range=0.2,
    height_shift_range=0.2,
    shear_range=0.2,
    zoom_range=0.2,
    horizontal_flip=True,
    fill_mode='nearest'
    
)
batch_size = 20
target_size = (224,224)

train_generator = train_imagenerator.flow_from_directory(
        train_dir, #parent directory must be specified # or use train_dir
        target_size = target_size, # All images will be resized to (224,224)
        batch_size=batch_size,
        class_mode='categorical' # since we need binary labels(0,1) and we will use binary_crossentropy

) 
```

```python
train_generator
```

```python
val_imagenerator = ImageDataGenerator(rescale=1/255.)
val_generator = val_imagenerator.flow_from_directory(
        validate_dir, #parent directory must be specified
        target_size = target_size, # All images will be resized to (180,180)
        batch_size=batch_size,
        class_mode='categorical' # since we need binary labels(0,1) and we will use binary_crossentropy
) 
```

```python
from tensorflow.keras.applications import ResNet50
from tensorflow.keras.layers import Dense
from tensorflow.keras.models import Model

# Number of butterfly species
num_classes = 100  # Replace with the actual number of classes

# Load the pre-trained ResNet50 model with top classification layers
pretrained_resnet50 = ResNet50(weights='imagenet', include_top=True)

# Remove the existing top classification layer
pretrained_resnet50.layers.pop()

# Add your custom classification layer
predictions = Dense(num_classes, activation='softmax')(pretrained_resnet50.layers[-1].output)

# Create the new model
model = Model(inputs=pretrained_resnet50.input, outputs=predictions)
```

```python
model.compile(optimizer='adam', loss='categorical_crossentropy', metrics=['accuracy'])
```

```python
epochs = 10  # Adjust the number of epochs as needed

history = model.fit(
    train_generator,
    steps_per_epoch=train_generator.samples // batch_size,
    epochs=epochs,
    validation_data=val_generator,
    validation_steps=val_generator.samples // batch_size
)
```

```python
num_classes = 100
from tensorflow.keras.applications import MobileNetV2

# Load MobileNetV2 model without top layers
base_mobilenetv2 = MobileNetV2(input_shape=(224, 224, 3), include_top=False, weights='imagenet')

# Add custom classification layers
x = base_mobilenetv2.output
x = keras.layers.GlobalAveragePooling2D()(x)
x = keras.layers.Dense(512, activation='relu')(x)
predictions = keras.layers.Dense(num_classes, activation='softmax')(x)

# Create the modified model
model = keras.models.Model(inputs=base_mobilenetv2.input, outputs=predictions)

# Compile the model
model.compile(optimizer='adam', loss='categorical_crossentropy', metrics=['accuracy'])

# Train the model
epochs = 10  # Adjust as needed

history = model.fit(
    train_generator,
    steps_per_epoch=train_generator.samples // batch_size,
    epochs=epochs,
    validation_data=val_generator,
    validation_steps=val_generator.samples // batch_size
)
```

```python
image_files = sorted([f for f in os.listdir(test_dir)], key=lambda x: int(re.search(r"(\d+)\.[a-z]+$", x).group(1)))
```

```python
image_files
```

```python
for image_file in image_files:
    # Load and preprocess the image
    image_path = os.path.join(test_dir, image_file)
    image = cv2.imread(image_path)
    image = cv2.cvtColor(image, cv2.COLOR_BGR2RGB)
    target_size = (224, 224)  # ResNet50 input size
    image = cv2.resize(image, target_size)
    preprocessed_image = preprocess_input(image)
    
    # Make predictions
    predictions = pretrained_resnet50.predict(tf.expand_dims(preprocessed_image, axis=0))
    decoded_predictions = decode_predictions(predictions, top=5)[0]
    
    print(f"Predictions for {image_file}:")
    for label, description, score in decoded_predictions:
        print(f"{description}: {score:.2f}")
```
