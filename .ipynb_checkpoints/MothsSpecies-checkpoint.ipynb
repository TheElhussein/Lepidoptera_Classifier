```python
import os
import pandas as pd
from tensorflow.keras.preprocessing.image import ImageDataGenerator, load_img
from PIL import Image

def is_valid_image(file_path):
    try:
        # Open and check the image
        with Image.open(file_path) as img:
            img.verify()
        return True
    except Exception as e:
        return False

folder_path = r'D:\Studying\ApplAi_CV\Assignments\third\train'
validate_dir = r'D:\Studying\ApplAi_CV\Assignments\third\valid'

# Create a list to store data for training and validation sets
train_data = []
val_data = []

# Create a mapping from label names (words) to numerical indices
label_to_index = {}
index = 0

# Process training dataset
for category in os.listdir(folder_path):
    category_path = os.path.join(folder_path, category)
    if os.path.isdir(category_path):
        label_to_index[category] = index
        index += 1
        for filename in os.listdir(category_path):
            image_path = os.path.join(category_path, filename)
            if is_valid_image(image_path):
                train_data.append({'filename': image_path, 'label': label_to_index[category]})
index = 0
# Process validation dataset
for category in os.listdir(validate_dir):
    category_path = os.path.join(validate_dir, category)
    if os.path.isdir(category_path):
        label_to_index[category] = index
        index += 1
        for filename in os.listdir(category_path):
            image_path = os.path.join(category_path, filename)
            if is_valid_image(image_path):
                val_data.append({'filename': image_path, 'label': label_to_index[category]})

# Create DataFrames for training and validation
train_df = pd.DataFrame(train_data)

val_df = pd.DataFrame(val_data)

# Define data generators and continue with your previous code
```

```python
filenames = os.listdir(folder_path)
data = []

for name in filenames:
    category_path = os.path.join(folder_path, name)
    if os.path.isdir(category_path):
        category_label = name
        for sub in os.listdir(category_path):
            data.append({'filename': os.path.join(category_path,sub), 'label': category_label})
            
train_df = pd.DataFrame(data)
```

```python
train_imagenerator = ImageDataGenerator(
     rescale=1/255.,
    zoom_range=0.2,
    horizontal_flip=True,
    fill_mode='nearest'
)
batch_size = 20
target_size = (224,224)

train_generator = train_imagenerator.flow_from_dataframe(
        train_df, 
        x_col='filename',
        y_col='label',
        target_size = target_size, 
        batch_size=batch_size,
        class_mode='raw'

) 
```

```python
filenames = os.listdir(validate_dir)
data = []

for name in filenames:
    category_path = os.path.join(validate_dir, name)
    if os.path.isdir(category_path):
        category_label = name
        for sub in os.listdir(category_path):
            data.append({'filename': os.path.join(category_path,sub), 'label': category_label})
            
val_df = pd.DataFrame(data)
```

```python
val_imagenerator = ImageDataGenerator(rescale=1/255)
val_generator = val_imagenerator.flow_from_dataframe(
        val_df,
        X_col='filename',
        y_col='label',
        target_size = target_size, 
        batch_size=batch_size,
        class_mode='raw' 
) 
```

```python
species = sorted([f for f in os.listdir(folder_path)]) 

image_files = [imagee for imagee in os.listdir(test_dir)]

sorted_image_files = sorted(image_files, key=lambda x: int(re.search(r'\d+', x).group()))
```

```python
input_shape = (224,224,3)

base_model = MobileNetV2(
            input_shape=input_shape,
            include_top=False,
            weights='imagenet')

#2. Locks all the base model's weights
for layer in base_model.layers:
    layer.trainable = False

#3. Adds a few layers to the end of the model
x = Flatten()(base_model.output)
x = Dense(512, activation='relu')(x)
x = Dropout(0.3)(x)
x = Dense(100, activation='softmax')(x)


transfer_model = tf.keras.Model(base_model.input, x)

optimizer = Adam(learning_rate=0.001)  # Adjust the learning rate as needed

transfer_model.compile(loss='sparse_categorical_crossentropy',
              optimizer=Adam(),
              metrics=['accuracy'])
```

```python
history = transfer_model.fit(
    train_generator,
    epochs=20,
    validation_data=val_generator,
)
```

```python
predictions = []

for image_file in sorted_image_files:
    full_image_path = os.path.join(test_dir, image_file)  

    img = cv2.imread(full_image_path)  
    img = cv2.resize(img, (224, 224))
    img_array = img_to_array(img)

    img_array /= 255.0

    
    img_array = img_array[np.newaxis, ...]

    prediction = transfer_model.predict(img_array)
    predicted_class_index = np.argmax(prediction)
    predicted_class_label = species[predicted_class_index]  

    predictions.append((image_file, predicted_class_label))
```
